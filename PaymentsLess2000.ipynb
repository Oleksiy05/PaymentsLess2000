```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as pl
sns.set(style="whitegrid")
```

```python
df = pd.read_excel('VPZ1.xlsx', encoding="cp1251")
```

```python
df1 = df.loc[(df['POPULATION'] <= 2000) & (df['POPULATION'] > 0) & (df['AVT_NUM'] == 0) & (df['CITY_TYPE'] == 'село')]
```

```python
#Сільске населення 13'093'100
#Кільксть домогосподарств 4'844'200
#Відношення 0.37
```

```python
df1.head()
```

```python
df1['HOUSEHOLD'] = df1['POPULATION']*0.37*0.8
df1['EXPECTED_QNT'] = df1['HOUSEHOLD']*3.3
df1 = df1[['ID_TECH','DIR','DIST','CITY','CITY_TYPE','AVT_NUM','POPULATION','PAY_QNT','PAY_AMOUNT','HOUSEHOLD','EXPECTED_QNT']]
df1['QNT_SHARE%'] = df1['PAY_QNT']/df1['EXPECTED_QNT']*100
```

```python
df1['SHARE_IMPROVE%']=df1["QNT_SHARE%"].mean()-df1['QNT_SHARE%']
```

```python
df1.head()
```

```python
df2 = df1.loc[df1['QNT_SHARE%'] >= 100]
```

```python
df2.shape
```

```python
df2
```

```python
df1["QNT_SHARE%"].mean()
```

```python
sns.set_style('ticks')
fig, ax = pl.subplots()
fig.set_size_inches(16, 8)
sns.barplot(x="DIR", y="QNT_SHARE%", palette=["m", "g"], data=df1, ax=ax)
ax.axhline(30.5, ls='--')
pl.xticks(rotation=90)
```

```python
df_g = df1.groupby('DIR')['QNT_SHARE%','SHARE_IMPROVE%'].mean()
```

```python
df_g
```

```python
df_g["SHARE_IMPROVE%"].mean()
```

```python
df_drop = df_g.loc[df_g['SHARE_IMPROVE%'] > 0]
```

```python
df_drop.shape
```

```python
df_drop["SHARE_IMPROVE%"].mean()
```

```python
df_gg = df1.groupby('DIR')['PAY_QNT','EXPECTED_QNT'].sum()
```

```python
df_gg
```

```python
df_m = pd.merge(df_drop, df_gg, how='left', on='DIR')
```

```python
#Norm 30.5% + 15%
df_m['NORM_QNT'] = df_m['EXPECTED_QNT']*0.4
df_m['QNT_TO_IMPROVE'] = df_m['NORM_QNT'] - df_m['PAY_QNT']
```

EXPECTED_QNT за кількістю домогосподарств до населення<br>
NORM_QNT 30.5% + 15% середня частка до максимальної кількості<br>
QNT_TO_IMPROVE кількість до середнього паказника + 15% росту
